```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# Define the custom font
font_path = 'C:/Users/ibsan/Desktop/TenX/week-5/data/fonts/AbyssinicaSIL-Regular.ttf'
custom_font = fm.FontProperties(fname=font_path)

# Get all tokens and count their frequency
all_tokens = cleaned_data['Tokens'].explode()
word_freq = all_tokens.value_counts().head(20)

# Plot word frequency
plt.figure(figsize=(12, 6))
word_freq.plot.bar(color='skyblue')

# Set the custom font for x-axis labels
plt.xticks(range(len(word_freq)), word_freq.index, rotation=45, fontproperties=custom_font, fontsize=12)
plt.title('Top 20 Most Common Words', fontproperties=custom_font)
plt.xlabel('Words', fontproperties=custom_font)
plt.ylabel('Frequency', fontproperties=custom_font)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# Define the custom font
font_path = 'C:/Users/ibsan/Desktop/TenX/week-5/data/fonts/AbyssinicaSIL-Regular.ttf'
custom_font = fm.FontProperties(fname=font_path)

# Calculate message lengths
cleaned_data['Message_Length'] = cleaned_data['Tokens'].str.len()

# Plot message length distribution
plt.figure(figsize=(10, 6))
cleaned_data['Message_Length'].plot.hist(bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Message Lengths', fontproperties=custom_font)
plt.xlabel('Number of Words', fontproperties=custom_font)
plt.ylabel('Frequency', fontproperties=custom_font)
plt.xticks(fontproperties=custom_font)
plt.yticks(fontproperties=custom_font)
plt.tight_layout()
plt.show()
```

```python
# Ensure the 'Date' column is in datetime format
cleaned_data['Date'] = pd.to_datetime(cleaned_data['Date'], errors='coerce')

# Drop rows with invalid dates
cleaned_data = cleaned_data.dropna(subset=['Date'])

# Resample by day and plot
time_series = cleaned_data.set_index('Date').resample('D').size()

plt.figure(figsize=(12, 6))
time_series.plot(color='orange')
plt.title('Messages Over Time')
plt.xlabel('Date')
plt.ylabel('Message Count')
plt.grid(True)
plt.show()
```
